## Abstract

Drug overdose mortality has become a pressing public health crisis in the United States, with significant variation across counties and regions. Economic conditions, such as unemployment rates, per-capita income, and regional economic activity, have been suggested as potential predictors of substance abuse and overdose deaths. This study investigates the extent to which county-level economic indicators in California from 2010–2019 can predict drug overdose mortality rates, and whether changes in these indicators signal future increases in overdoses.

To address this question, we compiled and merged multiple datasets, including CDC WONDER overdose mortality data, U.S. Bureau of Economic Analysis personal income data, Bureau of Labor Statistics unemployment statistics, and Fair Market Rent data. We standardized county and year identifiers, calculated per-capita overdose rates, and created lagged economic variables to examine whether economic shocks preceded increases in overdose mortality. Exploratory data analysis included descriptive statistics, correlation assessments, and linear regression models, both pooled and per-county, with outlier removal to ensure robust estimation.

Our analysis revealed that higher unemployment rates and lower per-capita income are generally associated with higher overdose mortality rates. In particular, unemployment increases were most strongly associated with overdose spikes one to two years later, supporting the hypothesized lag effect of economic stress on public health outcomes. Fair market rents and labor force size also showed moderate predictive power, highlighting the influence of local economic pressure beyond simple income measures.

These findings suggest that monitoring county-level economic indicators can provide early warnings for regions at risk of rising drug overdose deaths. Policymakers and public health officials may use such insights to target preventative interventions, allocate resources, and mitigate the societal impact of economic distress on substance abuse. Overall, our study underscores the significant interplay between economic conditions and public health, emphasizing the value of integrating socio-economic data into predictive models for overdose mortality.

## Hypothesis


**Hypothesis**: Higher county-level unemployment rates and lower per-capita personal income will be able to accurately predict increases in drug overdose mortality rates per 100,000 residents in California counties from 2018–2023. Furthermore, increases in unemployment will predict increases in overdose mortality rates with a lag of one to two years across California counties.
<br>

**Null Hypothesis**: County-level unemployment rates, per-capita personal income, and their lagged values are not significantly associated with drug overdose mortality rates in California counties from 2018–2023.
<br>

**Rationale**: Economic distress has been linked with increased substance use and overdose mortality due to reduced social security, financial strain, and decreased access to healthcare. The proxies we decided to look at include: Unemployment rate, which represents short-term economic shock, per-capita income, and regional economic activity, which both capture more long-term financial stress. Consequently, counties experiencing higher unemployment and lower income levels are more likely to exhibit higher rates of overdose mortality. Furthermore, economic shocks may not immediately translate into mortality, suggesting that there is a significant lag from when unemployment increases to a spike in drug overdose mortalities by one or more years.

### Dataset #1 California Drug Overdose Deaths by County

#### Dataset Description 
This dataset contains drug overdose mortality data for counties in California obtained from CDC WONDER. Each row represents the number of overdose deaths classified using **ICD-10 codes**. ICD-10 is an international alphanumeric classification system used to identify causes of death. This analysis focuses on substance-specific ICD-10 codes including T40.0–T40.5, T42.3–T42.4, and T43.6–T43.8, which correspond to commonly observed drugs such as opium, heroin, other opioids, synthetic opioids, cocaine, psychostimulants, and benzodiazepines.

The dataset also includes broader poisoning intent codes with the columns *Underlying Cause of Death* and *Multiple Cause of Death* variables. Codes X40–X44 represent accidental drug poisoning, X60–X64 represent intentional self-poisoning, and Y10–Y14 represent poisoning of undetermined intent. The Underlying Cause of Death (UCD) identifies the primary cause initiating the fatal event, while the Multiple Cause of Death (MCD) provides additional contributing substances involved in the death.

Geographic information is provided through the *County* and *County Code* variables, which identify the reporting county. These are captured by the Year and Year Code variables, which indicate the calendar year in which deaths occurred. CDC WONDER provided the reported death rates with the range of 2018 to 2023.

Outcome measures include *Deaths* and *Population*, representing the number of recorded deaths and the county population for that year. The dataset also reports a *Crude Rate*, defined as the number of deaths per 100,000 population. When death counts are small (typically below 20), CDC WONDER labels the crude rate as “Unreliable,” indicating statistical instability in the rate estimate.

#### Dataset Limitations
One of the potential problems in this data set in this dataset would be the statistically small crude rates. Some crude rates are miniscule that they are labeled as "Unreliable". This can become a potential limitation for analysis when working with the crude rate data. Another problem is the short annual range of 2018 - 2023. This small range makes it difficult to measure drastic spikes in overdose death overtime. The UCD and MCD are redundant variables because the ICD-10 codes already help group the drugs.

##### Understanding Durg Overdose Death Rate Variability between County

```python
# Pivot the data to get 2018 and 2023 side-by-side per county
# Assume 'Deaths' column is already numeric and NaNs are 0
county_growth = df_X40s.pivot_table(index='County', columns='Year', values='Deaths', aggfunc='sum')
df_X40_agg = df_X40s.pivot_table(index='County', columns='Year', values='Deaths', aggfunc='sum')

#  Calculate the Percentage Change
county_growth['pct_change'] = ((county_growth[2023] - county_growth[2018]) / county_growth[2018]) * 100

hotspots = county_growth.sort_values(by='pct_change', ascending=False)

print("Top 10 Counties with Highest Overdose Growth (2018-2023):")
print(hotspots[['pct_change']].head(10))
```

##### Removing Outlier counties

Given the 1433% increase in drug overdose death rate in Santa Cruz county and others, it is important that we remove the outliers.

```python
# Reshape df_X40_agg
df_long = df_X40_agg.reset_index()

# Melt the dataframe
# 'id_vars' are the columns you want to keep as they are
# 'var_name' will be the name of the new Year column
# 'value_name' will be the name of your deaths column
df_long = df_long.melt(
    id_vars=['County'], 
    var_name='Year', 
    value_name='total_overdose_deaths'
)

# Sort to keep counties together (optional but helpful)
df_long = df_long.sort_values(['County', 'Year']).reset_index(drop=True)
df_long
```

This model uses simple linear regression to explore whether socioeconomic variables such as income, labor force, fair market rent, and unemployment rate have a visible relationship with overdose deaths per 100,000 across California counties from 2018–2023. The goal is to identify which predictors show the strongest across all counties.

```python
# Code for First Pass Linear Regression
"""
CA County Linear Regression (2018–2023) — Outliers Removed
============================================================
Outcome (Y): overdose deaths per 100K population
Predictors (X): personal_income, labor_force, fmr_2br, unemployment_rate

Outlier removal: IQR method (1.5× IQR) applied independently to both
X and Y for each predictor graph (per-predictor cleaning).

Usage:
    pip install pandas scikit-learn matplotlib seaborn
    python ca_linear_regression_v3.py
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# ── 1. Load & prep ────────────────────────────────────────────────────────────

df = pd.read_csv("CA_merged_2018_2023.csv")
df["Year"] = df["Year"].astype(int)
df["overdose_per_100k"] = df["total_overdose_deaths"] / df["population"] * 100_000
df = df.dropna(subset=["overdose_per_100k"])

PREDICTORS = {
    "Personal Income ($)":      "personal_income",
    "Labor Force":               "labor_force",
    "Fair Market Rent 2BR ($)": "fmr_2br",
    "Unemployment Rate (%)":    "unemployment_rate",
}

OUTCOME_LABEL = "Overdose Deaths per 100K"
OUTCOME_COL   = "overdose_per_100k"


# ── 2. Helpers ────────────────────────────────────────────────────────────────

def remove_iqr_outliers(data, cols):
    """Remove rows where any of `cols` fall outside 1.5×IQR fences."""
    mask = pd.Series(True, index=data.index)
    for col in cols:
        q1, q3 = data[col].quantile(0.25), data[col].quantile(0.75)
        iqr = q3 - q1
        mask &= (data[col] >= q1 - 1.5 * iqr) & (data[col] <= q3 + 1.5 * iqr)
    return data[mask]

def fit_regression(X, y):
    model = LinearRegression().fit(X, y)
    preds = model.predict(X)
    return model, preds, r2_score(y, preds)


# ── 3. Pooled regressions (with and without outliers) ─────────────────────────

print("\n" + "="*60)
print(f"POOLED REGRESSIONS — Y = {OUTCOME_LABEL}")
print(f"IQR outlier removal applied per predictor (both X and Y)")
print("="*60)

pooled_results   = {}   # cleaned
pooled_raw       = {}   # original (for comparison)

for label, col in PREDICTORS.items():
    sub_raw   = df.dropna(subset=[col])
    sub_clean = remove_iqr_outliers(sub_raw, [col, OUTCOME_COL])

    n_removed = len(sub_raw) - len(sub_clean)

    for tag, sub, store in [("raw", sub_raw, pooled_raw),
                             ("clean", sub_clean, pooled_results)]:
        X = sub[[col]].values
        y = sub[OUTCOME_COL].values
        model, _, r2 = fit_regression(X, y)
        store[label] = {
            "col": col, "slope": model.coef_[0],
            "intercept": model.intercept_, "r2": r2,
            "model": model, "sub": sub,
        }

    print(f"\n{label}")
    print(f"  n (raw / clean) : {len(sub_raw)} / {len(sub_clean)}  "
          f"({n_removed} outliers removed)")
    print(f"  R² before       : {pooled_raw[label]['r2']:.4f}")
    print(f"  R² after        : {pooled_results[label]['r2']:.4f}")
    direction = "↑ positive" if pooled_results[label]['slope'] > 0 else "↓ negative"
    print(f"  Direction       : {direction}")


# ── 4. Per-county regressions (cleaned data) ──────────────────────────────────

print("\n" + "="*60)
print("PER-COUNTY REGRESSIONS (cleaned data)")
print("="*60)

county_results = {label: [] for label in PREDICTORS}

for label, col in PREDICTORS.items():
    sub_clean = pooled_results[label]["sub"]
    for county, grp in sub_clean.groupby("county"):
        grp = grp.sort_values("Year")
        if len(grp) < 3:
            continue
        X = grp[[col]].values
        y = grp[OUTCOME_COL].values
        model, _, r2 = fit_regression(X, y)
        county_results[label].append({
            "county": county, "slope": model.coef_[0],
            "intercept": model.intercept_, "r2": r2,
        })

for label, rows in county_results.items():
    cdf = pd.DataFrame(rows).sort_values("r2", ascending=False)
    print(f"\n── {label}  (top 5 counties by R²) ──")
    print(cdf[["county", "slope", "r2"]].head(5).to_string(index=False))


# ── 5. Regression scatter plots ───────────────────────────────────────────────

sns.set_theme(style="whitegrid", palette="muted")
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle(
    f"CA County: {OUTCOME_LABEL} vs. Predictors\n"
    f"(Outliers removed via IQR · per-predictor cleaning · 2018–2023)",
    fontsize=13, fontweight="bold"
)

scatter_ref = None
for ax, (label, col) in zip(axes.flat, PREDICTORS.items()):
    raw   = pooled_raw[label]
    clean = pooled_results[label]
    sub   = clean["sub"]

    # 5th–95th percentile axis limits for zoom (on cleaned data)
    x_lo, x_hi = sub[col].quantile(0.05), sub[col].quantile(0.95)
    y_lo, y_hi = sub[OUTCOME_COL].quantile(0.05), sub[OUTCOME_COL].quantile(0.95)
    x_pad = (x_hi - x_lo) * 0.04
    y_pad = (y_hi - y_lo) * 0.04

    # Only plot points that fall within the zoomed window
    in_view = (
        sub[col].between(x_lo - x_pad, x_hi + x_pad) &
        sub[OUTCOME_COL].between(y_lo - y_pad, y_hi + y_pad)
    )
    sub_view = sub[in_view]

    scatter_ref = ax.scatter(
        sub_view[col], sub_view[OUTCOME_COL],
        c=sub_view["Year"], cmap="coolwarm", alpha=0.65, s=26, linewidths=0,
        label=f"n={len(sub_view)} observations", zorder=2
    )

    # Regression line clipped to zoom window
    x_range = np.linspace(x_lo - x_pad, x_hi + x_pad, 200).reshape(-1, 1)
    ax.plot(x_range, clean["model"].predict(x_range),
            color="crimson", linewidth=2.2,
            label=f"Fit  R²={clean['r2']:.4f}", zorder=3)

    ax.set_xlim(x_lo - x_pad, x_hi + x_pad)
    ax.set_ylim(y_lo - y_pad, y_hi + y_pad)

    ax.set_title(label, fontsize=11, fontweight="bold")
    ax.set_xlabel(label, fontsize=9)
    ax.set_ylabel(OUTCOME_LABEL, fontsize=9)
    ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f"{v:,.0f}"))
    ax.legend(fontsize=7.5)

plt.colorbar(scatter_ref, ax=axes.flat[-1], label="Year", shrink=0.8)
plt.tight_layout()
plt.savefig("ca_regression_plots_clean.png", dpi=150, bbox_inches="tight")
print("\n  Plot saved → ca_regression_plots_clean.png")


# ── 6. Exploratory trend plots (cleaned) ──────────────────────────────────────

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle(
    f"CA County: {OUTCOME_LABEL} vs. Predictors — County Traces (Outliers Removed)",
    fontsize=13, fontweight="bold"
)

for ax, (label, col) in zip(axes.flat, PREDICTORS.items()):
    sub = pooled_results[label]["sub"]

    for county, grp in sub.groupby("county"):
        grp = grp.sort_values(col)
        ax.plot(grp[col], grp[OUTCOME_COL],
                color="gray", alpha=0.2, linewidth=0.8)

    x_range = np.linspace(sub[col].min(), sub[col].max(), 200).reshape(-1, 1)
    res = pooled_results[label]
    ax.plot(x_range, res["model"].predict(x_range),
            color="navy", linewidth=2.2,
            label=f"Pooled trend  R²={res['r2']:.4f}")

    # 5th–95th percentile zoom
    x_lo, x_hi = sub[col].quantile(0.05), sub[col].quantile(0.95)
    y_lo, y_hi = sub[OUTCOME_COL].quantile(0.05), sub[OUTCOME_COL].quantile(0.95)
    x_pad = (x_hi - x_lo) * 0.04
    y_pad = (y_hi - y_lo) * 0.04
    ax.set_xlim(x_lo - x_pad, x_hi + x_pad)
    ax.set_ylim(y_lo - y_pad, y_hi + y_pad)

    ax.set_title(label, fontsize=11, fontweight="bold")
    ax.set_xlabel(label, fontsize=9)
    ax.set_ylabel(OUTCOME_LABEL, fontsize=9)
    ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f"{v:,.0f}"))
    ax.legend(fontsize=8)

plt.tight_layout()
plt.savefig("ca_exploratory_trends_clean.png", dpi=150, bbox_inches="tight")
print("  EDA plot saved → ca_exploratory_trends_clean.png")


# ── 7. Export cleaned per-county results ──────────────────────────────────────

all_rows = []
for label, rows in county_results.items():
    for r in rows:
        all_rows.append({"parameter": label, **r})

pd.DataFrame(all_rows)[["parameter","county","slope","intercept","r2"]]\
    .to_csv("ca_regression_per_county_clean.csv", index=False)
print("  Per-county results saved → ca_regression_per_county_clean.csv")
```

```python
# 1. Define the variables
# Predictors (X): Labor, Personal Income, Fair Market Rent
# Target (y): Total Overdose Deaths (or population_od for the rate)
predictors = ['unemployment_rate', 'per_capita_income', 'fmr_2br']
target = 'total_overdose_deaths'

results = []

# 2. Group by County and Run Regression
for county, group in merge_raw.groupby('county'):
    # Clean data for this county (remove NaNs)
    data = group.dropna(subset=predictors + [target])
    
    # We need at least 4 data points for a multi-variable regression (3 predictors + 1)
    if len(data) >= 4:
        X = data[predictors]
        y = data[target]
        
        model = LinearRegression()
        model.fit(X, y)
        
        # Store results
        results.append({
            'County': county,
            'R2_Score': model.score(X, y),
            'Coef_Unemployment': model.coef_[0],
            'Coef_Income': model.coef_[1],
            'Coef_Rent': model.coef_[2],
            'Data_Points': len(data)
        })

# 3. Create a summary dataframe
regression_results = pd.DataFrame(results).sort_values(by='R2_Score', ascending=False)

# Display the top 10 counties where economic factors most strongly correlate with deaths
print("Top 10 Counties by R-Squared (Predictive Power):")
print(regression_results.head(10))
```

#### Final Model - Year Lagged OLS Pooled Linear Regression

This model uses a pooled OLS model to estimate a single statewide relationship between lagged unemployment, rent, and income and overdose death rates per 100,000 across all counties. By comparing predicted versus actual overdose rates and calculating mean absolute error (MAE) for each county, the analysis evaluates how reliably one “universal” economic model fits different local areas. The goal is to identify which counties are well explained by shared statewide economic patterns and show the highest MAE. 

```python
# # 1. Prepare Features (Using raw units as requested)
# merge_raw['income_per_capita'] = merge_raw['personal_income'] / merge_raw['population']
# merge_raw['overdose_rate_per_100k'] = (merge_raw['total_overdose_deaths'] / merge_raw['population']) * 100000

# # 2. Apply 1-Year Lags
# merge_raw['unemployment_lag1'] = merge_raw.groupby('county')['unemployment_rate'].shift(1)
# merge_raw['fmr_2br_lag1'] = merge_raw.groupby('county')['fmr_2br'].shift(1)
# merge_raw['income_lag1'] = merge_raw.groupby('county')['income_per_capita'].shift(1)

# 3. Clean and Set Panel Index (Pooled OLS still needs the index)
df_clean = df_sorted.dropna(subset=['lag_unemployment', 'lag_income', 'lag_rent', 'overdose_rate_per_100k']).copy()
df_panel = df_clean.set_index(['county', 'Year'])

# 4. Fit the Pooled OLS (The "Universal" Model)
# This model finds the best single line for the whole state
formula = 'overdose_rate_per_100k ~ 1 + lag_unemployment + lag_rent + lag_income'
pooled_model = PooledOLS.from_formula(formula, data=df_panel).fit()

# 5. Get Predictions directly from the fitted model
# Assign it to df_panel because they share the exact same MultiIndex
df_panel['predicted_rate'] = pooled_model.fitted_values

# (Optional) Clip negative predictions to 0, since death rates cannot be negative
df_panel['predicted_rate'] = df_panel['predicted_rate'].clip(lower=0)

# Calculate the error within the panel
df_panel['abs_error'] = abs(df_panel['overdose_rate_per_100k'] - df_panel['predicted_rate'])

# 6. Group by County to calculate the "Mean Prediction Error"
# We use level='county' because 'county' is now part of the MultiIndex, not a normal column
county_performance = df_panel.groupby(level='county').agg({
    'overdose_rate_per_100k': 'mean',
    'abs_error': 'mean'
}).rename(columns={'abs_error': 'MAE'})

# 7. Calculate "Accuracy Rate" 
county_performance['error_rate'] = (county_performance['MAE'] / county_performance['overdose_rate_per_100k']) * 100
state_mae = county_performance['MAE'].mean()

print(f"Statewide Mean Absolute Error: {state_mae:.2f} deaths per 100k")
print("\n--- Model Performance Summary ---")
print(county_performance[['MAE', 'error_rate']].sort_values('MAE').head(10))
```

```python
plt.figure(figsize=(14, 8))
sns.set_theme(style="whitegrid", context="talk")

# Sort counties by error for clear visualization
plot_df = county_performance.sort_values('MAE', ascending=True).reset_index()

# Create a bar chart of errors
sns.barplot(data=plot_df, x='MAE', y='county', palette='RdYlGn_r')

# Add a vertical line for the Statewide Average Error
plt.axvline(x=state_mae, color='black', linestyle='--', label=f'Statewide Avg Error: {state_mae:.1f}')

plt.title('Model Reliability by County: How well do economic signals predict deaths?\n(Mean Absolute Error per 100k)', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Mean Absolute Error (Deaths per 100k)', fontsize=14)
plt.ylabel('County', fontsize=12)
plt.legend()

plt.tight_layout()
plt.show()
```

## Conclusion and Discussion

This project investigated whether county-level economic indicators can be utilized to predict drug overdose mortality rates across California Counties from 2018 to 2023. Through our linear regression analysis, we have observed that there were some relationships between economic stressors and overdose mortality, but they were generally pretty weak as no individual factors had a huge correlation. Overall, unemployment rates are the biggest predictor of drug overdose death when including a 1-2 year lag between the unemployment rates and drug overdose deaths. The data showed significant variation between counties and years which made it hard to establish strong or reliable correlations between each economic variable and overdose rates. This may be due to external factors such as more populated counties, urban versus rural counties, and COVID-19. We were able to see the economic and health impacts of these external factors on our data visualization. 

When looking back at our previous research, our findings align with existing analysis on these topics. Previous studies that we looked at suggested that economic factors, in particular unemployment and financial instability, are correlated with higher substance use and overdose mortality. Our analysis and prediction model do show the same but the pattern was not consistent enough to support a strong predictive claim. This suggests that the economic parameters we looked at do influence overdose trends but we need to look beyond just economic factors if we want to create an accurate prediction model. Our work extends on already existing research by concentrating only on macroeconomic predictors to explain public health issues.

Overall, our results lead us to a broader conclusion which is that overdose mortality is influenced by a combination of factors that go beyond just the economic variables we looked at. Broadening our predictors to social and political factors as well will give us a more well-rounded prediction model which in turn would make it significantly more accurate. Some parameters that we think would be beneficial to look at include access to healthcare, prevalence of drugs, local drug policy differences. In addition to this, we saw that there was high variation in overdose rates across the counties which indicates local context does impact the model so statewide economic indicators may not capture all the necessary factors. 

There were some contradictory or unclear patterns in the data, for example certain counties with higher income levels still had relatively high overdose mortality rates which goes against our hypothesis that higher income means less economic stress and therefore less need for substance abuse. Moreover, fluctuations in overdose rates over time were not aligned in years with changes in the economic predictors. However, we had anticipated this and therefore we looked at time lags. Substance abuse may be prevalent at the same time as economic changes however drug overdose mortality may take time. Furthermore, we looked at multiple economic variables but they are very commonly interlinked, so it is hard to isolate individual impacts. 

There are several limitations to our analysis and prediction model that are important to address. Firstly, the datasets themselves lack scope because it aggregates data at county level which may hide within county variations. Secondly, we looked at data from 2018-2023 which may not be enough time to establish a definitive pattern and on top of that covid years were included which definitely changed the trend. Furthermore, the analysis we did is mostly correlational and cross-sectional which may mean that we can’t directly determine causation or relationships. Measurement biases were also present in how overdose deaths or economic indicators are reported by each county which may have further influenced the results we obtained. Analyses like these risk oversimplifying a very complex issue like drug overdose. From an ethical point of view, it is important to look at structural barriers so that the analysis isn’t used to characterize specific communities or populations. 

If this research was to be extended, there are several steps we could take to improve the analysis and give a more versatile overview. The first step would be incorporating additional variables that also looked at social and cultural factors such as healthcare access, prescription drugs vs illegal use, and prevalence of drugs in each community. Investigating variables beyond just economic ones will improve the analysis because it will provide a more comprehensive understanding of all the factors impacting overdose mortality. Furthermore, another step we could take is expanding the dataset to include more years and also trying to isolate how covid changes the pattern that we saw. Using longitudinal data and methods could help further explore the 1 to 2 year lag that we observed. Finally, we could also look at more advanced correlation techniques such as multivariate regression and causal inference which would help us isolate each parameter and see each relationship in detail. 

To conclude, even though the analysis only showed a relatively weak correlation it highlights the complexity of the overdose crisis. The results suggest that economic factors such as unemployment, personal income, labor force and fair market rent all contribute to drug overdose mortality rates. However, in order to accurately predict it will require a multifaceted approach that looks at social, cultural and even political parameters.
